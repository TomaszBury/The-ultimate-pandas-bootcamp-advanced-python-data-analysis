```python
df = pd.DataFrame({
    'age':[12,13,14,16],
    'wieght':[41.1,34.5,83.2,90.1],
    'height':['1.72','1.74','1.91','1.54']
    })
df = df.astype(float)
df.loc[2,'wieght'] = np.nan
df.loc[1,:] = np.nan
df
```

```python
df.dropna(thresh=df.shape[1], axis=1)
```

```python
df['gender'] = ['M', "f", np.NaN, 'F']
df
```

```python
df.dropna(how='any', axis=0)
```

```python
df.dropna(how='any', axis=0, subset=['gender'])
```

```python
df.dropna(how='any', axis=0, subset=['age'])
```

```python
df.dropna(how='any', axis=1)
```

```python
df.dropna(how='all', axis=1)
```
