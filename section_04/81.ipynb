```python
import pandas as pd
```

```python
data_url = 'https://gist.githubusercontent.com/sh7ata/56976975b3c67caabb2efd80bb3a8467/raw/f1b7d2abf6770548edafee58ea15be71e2d28aac/Nutrition.csv'
```

```python
nutrition = pd.read_csv(data_url, index_col=[0])
nutrition.head(n=2)
```

Randomly selcet 10 food items and assign the resulting dataframe to new varaible called: nutr_mini

```python
nutr_mini = nutrition.sample(n=10)
nutr_mini
```

From nutr_mini, extract the total_fat and cholesterol clumns for all rows.

```python
col_cholesterol = nutr_mini.columns.get_loc('cholesterol')
col_cholesterol
```

```python
col_total_fat = nutr_mini.columns.get_loc('total_fat')
col_total_fat
```

```python
nutr_mini.iloc[:,[col_cholesterol,col_total_fat]]
```

```python
nutr_cholesterol_total_fat = nutr_mini.iloc[:,[col_cholesterol,col_total_fat]]
nutr_cholesterol_total_fat
```

```python
nutr_cholesterol_total_fat['Cholesterol_no_unit'] = nutr_cholesterol_total_fat.loc[:,'cholesterol'].str.split('mg',n=1,expand=True).iloc[:,0].copy()
nutr_cholesterol_total_fat['Total_fat_no_unit'] = nutr_cholesterol_total_fat.loc[:,'total_fat'].str.split('g',n=1,expand=True).iloc[:,0].copy()
nutr_cholesterol_total_fat
```

```python
nutr_cholesterol_total_fat.drop(axis=1, columns='cholesterol',inplace=True)
nutr_cholesterol_total_fat.drop(axis=1, columns='total_fat', inplace=True)
```
