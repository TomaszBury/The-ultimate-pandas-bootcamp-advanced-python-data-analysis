# 282. Bar. Charts  
A chart that displays a numerical value across levels of a categorical value.|

````python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

url = "https://andybek.com/pandas-marathon"

boston= pd.read_csv(url)
boston.head()
````

Marathon runners by age:

````python
boston.Age
````

````python
boston.Age.plot()

plt.show()
````

````python
boston.Age.value_counts()
````

````python
boston.Age.value_counts().plot()

plt.show()
````

````python
boston.Age.value_counts().plot(kind='bar', figsize=(12,6))

plt.show()
````

````python
ax = boston.Age.value_counts(sort=False).plot(kind='bar', figsize=(12,6))

ax.set_xlabel('Age Group')
ax.set_ylabel('# of runners')

plt.show()
````

````python
ax = boston.Age.value_counts(sort=False).sort_index().plot(kind='bar', figsize=(12,6))

ax.set_xlabel('Age Group')
ax.set_ylabel('# of runners')

plt.show()
````

````python
ax = boston.Age.value_counts(sort=False).sort_index().plot(kind='barh', figsize=(12,6))

ax.set_xlabel('Age Group')
ax.set_ylabel('# of runners')

plt.show()
````

Marathon runners by gender?

````python
boston['M/F'].value_counts(normalize=True).plot(kind='barh')

plt.show()
````

# 283 Pie Plots   
A plot that is used to show how a total amount i attributed between its component parts.

````python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

url_games = 'https://andybek.com/pandas-games'

games = pd.read_csv(url_games)
````

````python
games.head()
````

What is the market share for each platform in NA?

````python
games.NA_Sales
````

Does PS have a larger market share in JP?

````python
games.groupby('Platform').JP_Sales.sum().plot(kind='pie', figsize=(10,10))

plt.show()
````

# 285. Scatter plots   
Show the relationship between two numeric values

````python
url_sat = 'https://andybek.com/pandas-sat'
sat = pd.read_csv(url_sat)
sat.head(n=6)
````

## Is reading and writing more strongly associated than writing and math?

````python
mask_math = sat.loc[:,'SAT Section'] == 'Math'
mask_reading = sat.loc[:,'SAT Section'] == 'Reading'
mask_writing = sat.loc[:,'SAT Section'] == 'Writing'

math = sat.loc[mask_math].Score.reset_index(drop=True)
reading = sat.loc[mask_reading].Score.reset_index(drop=True)
writing = sat.loc[mask_writing].Score.reset_index(drop=True)
````

````python
scores = pd.concat([math, writing, reading], axis=1)
scores.columns = ['Math', 'Writing', 'Reading']
````

````python
scores.head(n=7)
````

````python
scores.info()
````

````python
scores.plot(kind='scatter', x='Math', y='Writing');
````

````python
math.corr(writing)
````

````python
scores.plot(kind='scatter', x='Reading', y='Writing');
````

````python
reading.corr(writing)
````
