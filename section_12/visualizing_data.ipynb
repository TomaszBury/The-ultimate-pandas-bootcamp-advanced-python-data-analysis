# 282. Bar. Charts  
A chart that displays a numerical value across levels of a categorical value.|

````python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

url = "https://andybek.com/pandas-marathon"

boston= pd.read_csv(url)
boston.head()
````

Marathon runners by age:

````python
boston.Age
````

````python
boston.Age.plot()

plt.show()
````

````python
boston.Age.value_counts()
````

````python
boston.Age.value_counts().plot()

plt.show()
````

````python
boston.Age.value_counts().plot(kind='bar', figsize=(12,6))

plt.show()
````

````python
ax = boston.Age.value_counts(sort=False).plot(kind='bar', figsize=(12,6))

ax.set_xlabel('Age Group')
ax.set_ylabel('# of runners')

plt.show()
````

````python
ax = boston.Age.value_counts(sort=False).sort_index().plot(kind='bar', figsize=(12,6))

ax.set_xlabel('Age Group')
ax.set_ylabel('# of runners')

plt.show()
````

````python
ax = boston.Age.value_counts(sort=False).sort_index().plot(kind='barh', figsize=(12,6))

ax.set_xlabel('Age Group')
ax.set_ylabel('# of runners')

plt.show()
````

Marathon runners by gender?

````python
boston['M/F'].value_counts(normalize=True).plot(kind='barh')

plt.show()
````

# 283 Pie Plots   
A plot that is used to show how a total amount i attributed between its component parts.

````python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

url_games = 'https://andybek.com/pandas-games'

games = pd.read_csv(url_games)
````

````python
games.head()
````

````python
games.Global_Sales.plot(kind='pie')
````

What is the market share for each platform in NA?

````python
games.NA_Sales
````

Does PS have a larger market share in JP?

````python
games.groupby('Platform').JP_Sales.sum().plot(kind='pie', figsize=(10,10))

plt.show()
````

````python
games.groupby('Platform').NA_Sales.sum().plot(kind='pie', figsize=(10,10))

plt.show()
````

````python
market_sales = ['NA_Sales','JP_Sales','Platform']

games.loc[:,market_sales].groupby('Platform').sum().plot(kind='pie',y='NA_Sales')

plt.show()
````

````python
market_sales = ['NA_Sales','JP_Sales','Platform']

games.loc[:,market_sales].groupby('Platform').sum().plot(kind='pie',subplots=True, figsize=(13,10))

plt.show()
````

````python
games.JP_Sales.sum()
````

# 284. Histograms   
Display the distribution of a numeric variable as a series of bars

````python
drinks_url = 'https://andybek.com/pandas-drinks'

drinks = pd.read_csv(drinks_url)

drinks.head()
````

````python
drinks.beer_servings.plot(kind='hist')

plt.show()
````

````python
ax = drinks.beer_servings.plot(kind='hist', bins=10)

ax.set_xlabel('Beer Servings')

plt.show()
````

````python
ax = drinks.beer_servings.plot(kind='hist', bins=5)

ax.set_xlabel('Beer Servings')

plt.show()
````

````python
ax = drinks.beer_servings.plot(kind='hist', bins=15)

ax.set_xlabel('Beer Servings')

plt.show()
````

````python
ax = drinks.beer_servings.plot(kind='hist', bins=35)

ax.set_xlabel('Beer Servings')

plt.show()
````

````python
ax = drinks.beer_servings.plot(kind='hist', bins=10, orientation='horizontal')

ax.set_xlabel('Beer Servings')

plt.show()
````
