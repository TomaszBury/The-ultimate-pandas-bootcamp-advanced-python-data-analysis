```python
import random
dates.iloc[random.randint(0,dates.shape[0]),0]
```

```python
def list_of_random_dates(n):
    random_dates = []
    for i in range(n):
        random_dates.append(dates.iloc[random.randint(0,dates.shape[0]-1),0])
    return random_dates
```

```python
list_dates = list_of_random_dates(5)
list_dates
```

```python
tech.loc[list_dates]
```

```python
tech.loc[(['2015-01-06','2015-01-07'],['FB','AMZN']),:]
```

```python
tech.loc[(list_of_random_dates(3), ['FB','AMZN']),:]
```
