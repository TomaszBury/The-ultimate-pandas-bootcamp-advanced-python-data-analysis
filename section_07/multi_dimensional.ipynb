```python
def list_of_random_dates(n):
    random_dates = []
    for i in range(n):
        random_dates.append(dates.iloc[random.randint(0,dates.shape[0]-1),0])
    return random_dates
```

```python
# 2. Select 10 days at randome from tech_df2, containing only AAPL price data.
```

```python
import random
from datetime import datetime, timedelta

min_year=2014
max_year=2019

start = datetime(min_year, 1, 1)
years = max_year - min_year+1
end = start + timedelta(days=365 * years)
list_days = []

for i in range(10):
    random_date = start + (end - start) * random.random()
    # print(f'{type(random_date.strftime("%Y-%m-%d"))} - {random_date.strftime("%Y-%m-%d")}')
    list_days.append(random_date.strftime("%Y-%m-%d"))
```

```python
list_days
```

```python
tech.loc[(list_days,'AAPL'),:]
```

```python
tech.loc[pd.IndexSlice[list_days,'AAPL'],:]
```

```python
tech.xs(list_days,level=0)
```

```python
tech.loc[list_days]
```
