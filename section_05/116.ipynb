```python
import pandas as pd
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.head(n=2)
```

```python
# reindex()

players_like = players.iloc[:4, :4]
```

```python
players_like
```

```python
# row  order: 2, 1, 3, 0
# column order: age, name, position, club

players_like.reindex(index=[2,1,3,0], columns=['age', 'name', 'position', 'club'])
```

```python
players_like
```

```python
players.reindex(index=[2,1,3,0], columns=['age', 'name', 'position', 'club'])
```

```python
players.reindex(index=[2,1,3,0])
```

```python
# ...get all the columns, and have them alphabetically ordered
players.reindex(index=[2,1,3,0]).sort_index(axis=1)
```

```python
players.columns.sort_values()
```

```python
index_players = [2,1,3,0]
players.reindex(index=index_players, columns=players.columns.sort_values())
```

```python
iter(players.columns)
```

```python
players.reindex(index=index_players, columns=sorted(players.columns))
```

```python
players.reindex(index=index_players, columns=sorted(players.columns)[:6])
```
