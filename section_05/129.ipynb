```python
import pandas as pd
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.head(n=2)
```

```python
players.loc[players.isna().values].drop_duplicates()
```

```python
players.fillna(method='ffill').loc[players.isna().values]
```

```python
players.loc[players.isna().values].drop_duplicates()
```

```python
players.fillna(method='ffill').iloc[[29,30,191,192,194,195]]
```

```python
na_players = players.loc[players.isna().values].drop_duplicates().index
na_players = list(na_players)
forward_na_players = []
for p in na_players:
    forward_na_players.append(p-1)
    forward_na_players.append(p)
forward_na_players
players.fillna(method='ffill', axis=0).iloc[forward_na_players]
```

```python
players.fillna(method='ffill', axis=1).loc[na_players]
```
