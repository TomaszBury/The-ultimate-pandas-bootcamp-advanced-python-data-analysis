```python
import pandas as pd
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.head(n=2)
```

```python
players.loc[players.duplicated()]
```

```python
players.market_value.mean()
```

```python
players.market_value.loc[players.market_value.duplicated()].mean()
```

```python
mask = players.duplicated(subset=None, keep='first')
players.loc[mask]
```

```python
players.market_value.loc[mask].mean()
```

```python
players.drop_duplicates(keep='first')
```

```python
players.drop_duplicates(keep='first').market_value.mean()
```

```python
players.market_value.loc[~mask].mean()
```

```python
players.drop_duplicates(keep=False).market_value.mean()
```
