```python
import pandas as pd
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.head(n=2)
```

# A sophisticated alternative

```python
# reindex to remove ROWs / COLUMNs
players.reindex()
```

```python
players.shape
```

```python
players.reindex().shape
```

```python
players.reindex(index=[0,3,9])
```

```python
unwanted_rows = [1,2,3,4]
players.reindex(index=set(players.index).difference(unwanted_rows))
```

```python
set(players.index)
```

```python
set(players.index).difference(unwanted_rows)
```

```python
unwanted_columns = ['name', 'position', 'position_cat']
players.reindex(columns=set(players.columns).difference(unwanted_rows))
```
