```python
import pandas as pd
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.head(n=2)
```

Find the players in our datset that meet these criteria:
1. They're Eneglish (Nationlity), and
2. Therir market value is more than twice the average market value in the league (market_value), and 
3. They either have more than 4,000 views (pate_views) or are a new signing (new_signing) but not both

```python
nationality = players.nationality == 'England'
market_value = players.market_value > (players.market_value.mean() * 2)
page_views = players.page_views > 4000
new_signing = players.new_signing == 1
```

```python
players.loc[nationality & market_value & (page_views ^ new_signing)]
```
