```python
import pandas as pd
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.head(n=2)
```

```python
players.loc[
    (players.position == 'LB') &
    (players.market_value >= 10) &
    (players.age <= 25) &
    ~(players.club.isin(['Totenham', 'Arsenal']))
]
```

```python
# new target: Arsenal right backs and chelsea goalkeepers

arsenal_player = players.club == 'Arsenal'
position_rb = players.position == 'RB'
chelsea_player_gk = (players.club == 'Chelsea') & (players.position == 'GK')
```
