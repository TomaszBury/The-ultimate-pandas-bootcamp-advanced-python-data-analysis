```python
import pandas as pd
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.head(n=2)
```

```python
players.loc[players.isna().values]
```

```python
players_no_na = players.fillna('some meaningful replacement value')
players_no_na.iloc[players.loc[players.isna().values].index]
```

```python
players.position.unique()
```

```python
players.fillna({
    'market_value': 100,
    'position': 'RM'
}).loc[players.loc[players.isna().values].index]
```

```python
players.fillna({
    'market_value': players.market_value.mean(),
    'position': 'RM'
}).loc[players.loc[players.isna().values].index]
```

```python
players.loc[players.loc[players.isna().values].index]
```

```python
players_no_na = players.dropna()
players_no_na.loc[players.loc[players.isna().values].index]
```
