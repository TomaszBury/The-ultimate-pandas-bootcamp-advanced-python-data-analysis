```python
import pandas as pd
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.head(n=2)
```

```python
players.duplicated()
```

```python
players.loc[players.duplicated()]
```

```python
# the subset param
# unique -> club, age, position, market_value
mask_subset=['club', 'age', 'position', 'market_value']
mask = players.duplicated(subset=mask_subset)
players.loc[mask]
```

```python
players.loc[mask].shape
```

```python
mask_keep = players.duplicated(subset=mask_subset, keep='last')
players.loc[mask_keep]
```

```python
players.loc[mask_keep].shape
```

```python
mask_keep = players.duplicated(subset=mask_subset, keep=False)
players.loc[mask_keep]
```

```python
players.loc[mask_keep].shape
```
