```python
import pandas as pd
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)
players.head(n=2)
```

```python
# Sort player in the player dataframe by age in ascending order. Who is the youngest footballer in the EPL?
players.sort_values(by='age', ascending=True).head(n=1)
```

```python
players.nsmallest(columns='age', n=1)
```

```python
# Set the club column as the index of the dataframe. then sort the dataframe index in alphabetical order. Make sure these changes are aplied to the underliying data frame and carry over to the next question.

players.set_index('club', inplace=True)
```

```python
players.sort_index(ascending=True, inplace=True)
```

```python
players.head(n=10)
```

```python
# Sort the dataframe values by club and market_value the club is alphabetical (Arsenal First) and the market_value is in descending order (with each team, the most valuable players first).
players.reset_index(inplace=True)
```

```python
players.sort_values(by=['club', 'market_value'], ascending=[True, False], inplace=True)
```
